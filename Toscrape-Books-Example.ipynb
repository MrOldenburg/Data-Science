```python
import re
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup

#this code scrapes data at torscrape for 1001 books, it gathers their title, price and star rating.

#store website link as string
myurl = 'http://books.toscrape.com/catalogue/category/books_1/page-1.html'

#iterations for each page
for i in range(1, 51):

    # create the URL for the current page
    url = myurl.replace('-1', "-" + str(i))

    # grab current page stored as uclient
    uClient = uReq(url)

    # read and close HTML
    page_html = uClient.read()
    uClient.close()

    # call BeautifulSoup for parsing
    page_soup = soup(page_html, "html.parser")

    # stores products in bookshelf
    bookshelf = page_soup.findAll(
        "li", {"class": "col-xs-6 col-sm-4 col-md-3 col-lg-3"})

    # create csv file
    filename = ("Books.csv")
    f = open(filename, "a")

    # checks if already written and writes headers
    if i == 1:
        headers = "Book title, Price, Stars\n"
        f.write(headers)

    for books in bookshelf:

        # collect titles
        book_title = books.h3.a["title"]
        book_title = re.sub(',', ' -', book_title)

        # collect prices
        book_price = books.findAll("p", {"class": "price_color"})
        price = book_price[0].text.strip()

        # collect the stars
        stars = books.find("p", "star-rating")
        starsfix = str(stars)
        #very crudely cleans it up
        striped = re.sub('[<>"/pclass=sta-atg^\s+|\s+$]','',starsfix)
        striped = re.sub('r-r','',striped)
        striped = re.sub('-r','', striped)
        striped = re.sub('on','', striped)
        striped = re.sub('n','', striped, count=1)
        striped = re.sub('i','', striped, count=1)
        striped = re.sub('ii','', striped)
        striped = striped[:-1]
        #prints the three outputs
        print("Title of the book :" + book_title)
        print("Price of the book :" + price)
        print("Star Rating : " + striped)

        # write the book information to the CSV file
        f.write(book_title + "," + price + "," + striped + "\n")

    # close the CSV file after each page has been scraped
    f.close()
```
